```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read train.csv file into a DataFrame
df_credit = pd.read_csv('./train.csv')

df_credit
```

```python
# Remove underscores and convert "Age" column to float
df_credit['Age'] = df_credit['Age'].str.replace('_', '').astype(float)
```

```python
# Remove underscores and convert "Annual_Income" column to float
df_credit['Annual_Income'] = df_credit['Annual_Income'].str.replace('_', '').astype(float)


# Remove underscores from "Num_of_Loan" column
df_credit["Num_of_Loan"] = df_credit["Num_of_Loan"].str.replace('_', '').astype(float)

# Remove underscores from "Num_of_Delayed_Payment" column
df_credit["Num_of_Delayed_Payment"] = df_credit["Num_of_Delayed_Payment"].str.replace('_', '').astype(float)



# Remove underscores from "Outstanding_Debt" column
df_credit["Outstanding_Debt"] = df_credit["Outstanding_Debt"].str.replace('_', '').astype(float)

# Remove underscores from "Amount_invested_monthly" column
df_credit["Amount_invested_monthly"] = df_credit["Amount_invested_monthly"].str.replace('_', '').astype(float)

# Remove underscores from "Monthly_Balance" column
df_credit["Monthly_Balance"] = df_credit["Monthly_Balance"].str.replace('_', '').astype(float)
```

```python
# Remove underscores from "Changed_Credit_Limit" column
df_credit["Changed_Credit_Limit"] = df_credit["Changed_Credit_Limit"].str.replace('_', '')

# Replace empty strings with NaN
df_credit["Changed_Credit_Limit"].replace('', np.nan, inplace=True)

# Convert "Changed_Credit_Limit" column to float
df_credit["Changed_Credit_Limit"] = df_credit["Changed_Credit_Limit"].astype(float)

```

```python
# Function to convert age strings to months
def convert_to_months(age_str):
    if pd.isnull(age_str):  # Check if value is NaN
        return np.nan
    else:
        years, months = age_str.split(' Years and ')
        years = int(years)
        months = int(months.split(' Months')[0])
        total_months = years * 12 + months
        return total_months

# Apply function to convert to months
df_credit['Credit_History_Age_Months'] = df_credit['Credit_History_Age'].apply(convert_to_months)

df_credit
```

```python
# Write the DataFrame to a CSV file
df_credit.to_csv('modified_df_credit_train.csv', index=False)
```
